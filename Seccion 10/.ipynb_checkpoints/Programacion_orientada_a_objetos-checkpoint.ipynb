```python
class DataSet:
    tipo_dato = 'numerico'

    def __init__(self, data):
        self.data = data

    def media(self):
        """Devuelve la media de los datos"""
        return sum(self.data) / len(self.data)
    
    def mediana(self):
        sorted_data = sorted(self.data)
        n = len(sorted_data)
        midpoint = n // 2

        if n % 2 == 0:
            return (sorted_data[midpoint-1] + sorted_data[midpoint]) / 2
        else:
            return sorted_data[midpoint]
    
    def set_dato(self, nuevo_dato):
        self.data.append(nuevo_dato)

    
```

## Herencia y Polimorfismo


```python
class DataSet:
    tipo_dato = 'numerico'
    

    def __init__(self, data):
        self.data = data
        self.tamaño =  len(data)

    def media(self):
        """Devuelve la media de los datos"""
        return sum(self.data) / len(self.data)
    
```

```python
class DataSetPlus(DataSet):

    def desviacion_estandar(self):
        mu = self.media()
        return (sum((x-mu)**2 for x in self.data) / self.tamaño) ** 0.5
        
    
```

```python
datos = DataSet([3,5,8])
datosplus = DataSetPlus([3,5,8])
```

```python
datos.media()
```

```python
datosplus.media()
```

```python
datos.desviacion_estandar()
```

```python
datosplus.desviacion_estandar()
```

## Encapsulamiento y Abstraccion

```python
class DataSet:
    tipo_dato = 'numerico'

    def __init__(self, data):
        self._data = data         #Atributos privados
        self._tamaño =  len(data)

    def media(self):
        """Devuelve la media de los datos"""
        return sum(self._data) / len(self._data)
    
    def mediana(self):
        sorted_data = sorted(self._data)
        n = len(sorted_data)
        midpoint = n // 2

        if n % 2 == 0:
            return (sorted_data[midpoint-1] + sorted_data[midpoint]) / 2
        else:
            return sorted_data[midpoint]

    def varianza(self):
        mu = self.media()
        return (sum((x-mu)**2 for x in self._data) / self._tamaño) ** 0.5

    def obtener_data(self):
        return self._data.copy() #Devuelve una copia para proteger el atributo
    
    def set_dato(self, nuevo_dato):
        self.data.append(nuevo_dato)
```

```python
# Se usa una convencion para indicar que son meotos o atributos encapsulados
# En Python no se puede encapsular los atriutos y metodos, no hay una forma de restringir totalmente

datos = DataSet([4,3,7])
```

```python
datos.obtener_data()
```

```python
datos._data # Ejemplo que pese al encapsulamiento igual podemos acceder al atributo, a diferencia de otros lenguajes
```
