```python
# Dada una lista de ingresos mensuales, calcula media, mediana y desviación estándar.
ingresos = [3500,9000,4000,3000,3200,1800,1500,2000,1200,5000, 2000, 1500, 1800, 5000, 1500]

media = sum(ingresos)/len(ingresos)
print(f'La media de los ingresos es: S/ {media:.2f}')

mediana = 0
n = len(ingresos)
ingresos_orden = sorted(ingresos)

if n%2 == 0:
    mediana = ingresos_orden[n/2]
else:
    mediana = ingresos_orden[n//2-1]

print(n)
print(ingresos_orden)
print(f'La mediana de los ingresos es: {mediana} su indice es: {n//2-1}')

import math

suma = 0
for i in range(n):
    suma = suma + (ingresos[i] - media)**2

print(f'La desviacion estandar de los ingresos es: {math.sqrt(suma/(n-1)):.2f}')
```

```python
import statistics

def estadisticas(datos):
    datos_limpios = [x for x in datos if isinstance(x, (int, float))] # verifica que los datos sen int o float
    
    return {
        "media": statistics.mean(datos_limpios),
        "mediana": statistics.median(datos_limpios),
        "desviacion": statistics.stdev(datos_limpios)
    }
estadisticas(ingresos)
```

```python
def clean_values(data):
    data_clean = [x for x in data if isinstance(x, (int, float)) ]

    return data_clean


data = [1,2,3,5]

clean_values(data)
```

```python
def filtrar_mayores(lista, umbral):
    return [x for x in lista if x > umbral]

lista = [1,2,3,5]
filtrar_mayores(lista, 3)
```
