```python
# Consideraciones importantes y sintaxis en Python

def imprimir_numeros_pares_hasta_n(n):
    """
    Imprime todos los numeros pares desde 0 hasta en numero n.
    """
    for i in range(n + 1):
        if i % 2 == 0:
            print(i)
        else:
            print(f"{i} (impar)")

# Uso de la funcion
limite = 11
print(f"Numeros pares hasta {limite}:")
imprimir_numeros_pares_hasta_n(limite)


# Consideraciones finales
"""
En esta funcion, el bucle for  itera sobre los numeros desde 0 hasta n.
El condicional if-else verifica si el numero es par o impar
La identacion define los bloques de codigo dentro de la funcion
"""
```
