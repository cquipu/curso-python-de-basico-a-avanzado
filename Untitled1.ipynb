```python
visitas = [120, 150, 90, 200, 175, 80, 220]
```

```python
## total de visitas
total_visitas = sum(visitas)
print("total_visitas: ", total_visitas)
```

```python
## promedio de visitas
promedio_visitas = sum(visitas)/len(visitas)
print(f"promedio_visitas: {promedio_visitas:.2f}")
```

```python
## día con más visitas
max_visitas = max(visitas)
dia_max_visitas = visitas.index(max_visitas)
print(f"dia_max_visitas: {dia_max_visitas + 1}, numero_visitas: {max_visitas}")
```

```python
# Crea una función que reciba una lista de precios de productos y devuelva solo los precios mayores a 50.
def precios_mayor_50(precios):
    resultado = list()
    for precio in precios:
        if precio > 50:
            resultado.append(precio)
    return resultado


precio_frutas = [120, 150, 90, 20, 17, 80, 220]

print(precios_mayor_50(precio_frutas))
```

```python
"""
Crea una lista con 20 números aleatorios entre 1 y 100.
Luego:
- elimina los números pares
- ordena la lista
"""
import random

numeros = [random.randint(1, 100) for i in range(100)]
print(numeros)

resultado = list()
for num in numeros:
    if num % 2 != 0:
        resultado.append(num)

resultado.sort() # ordena los elementos de la lista

print(resultado)
```

```python
import os

def limpiar_pantalla():
    # 'nt' es para Windows, 'posix' para Linux/macOS
    os.system('cls' if os.name == 'nt' else 'clear')

productos = list()
producto = ""

status = True
while status:
    print("Elige una opcion")
    print("agregar producto: 1")
    print("eliminar producto: 2")
    print("mostrar carrito: 3")
    print("Salir carrito: s")

    opcion = int(input("ingresa"))

    match opcion:
        case 1:
            limpiar_pantalla()
            producto = input ("Escribe el nombre del Producto: ")
            productos.append(producto)
            print("Producto registrado!")        
        case 2:
            if len(productos) == 0:
                limpiar_pantalla()
                print("No hay productos ...")                
                continue
            else:
                limpiar_pantalla()
                print("Productos: ", productos)
                print("Escribe el producto a eliminar ...")
                producto = input("Nombre del producto: ")
                productos.remove(producto)
                print("Producto eliminado!")  
                print(productos)
        case 3:  
            limpiar_pantalla()
            print("Su carrito contiene ...")
            print("Carrito: ", productos)

        case _: 
            status = False

            if len(productos):                
                print("Gracias por su compra ...")
                print("Contenido del carrito: ", productos)
            else:
                print("Gracias por su compra, carrito vacio...")
            
```

```python
# 1. Nuestra lista de productos (esto es una lista de tuplas)
productos = [
    ("Laptop", 1200),
    ("Mouse", 500),
    ("Monitor", 300),
    ("Teclado", 75)
]

productos = dict(productos)
# 2. Definimos una función que "le diga" a Python qué mirar.
# Queremos el precio, que es el segundo elemento (índice 1).
def dame_el_precio(producto):
    return producto[1]

# 3. Usamos la función sorted() para crear una nueva lista ordenada.
# El secreto está en decirle: key=dame_el_precio
lista_ordenada = sorted(productos.items(), key=lambda item: item[1])
print(lista_ordenada)
# 4. Imprimimos el resultado para ver que funcionó
#print("Lista de productos ordenada por precio:")
#for item in lista_ordenada:
#    print(item)
```
