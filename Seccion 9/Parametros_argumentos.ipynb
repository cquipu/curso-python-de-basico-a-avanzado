```python
def saludar(nombre, edad):
    print(f'Hola mi nombre es {nombre} y mi edad es {edad}')
```

```python
saludar('Edson', 35)
```

```python
saludar(35, 'Edson')
```

```python
saludar(edad=26, nombre='Mateo')
```

```python
def saludar(nombre, ciudad="Madrid"):
    print(f'Hola mi nombre es {nombre} y vengo de {ciudad}')
```

```python
saludar('Jorge', 'Buenos Aires')
```

```python
saludar('Juan')
```
