```python
def operaciones_basicas(a,b):
    suma = a+b
    resta = a-b
    producto = a*b
    division = a / b
    return suma, resta, producto, division
```

```python
operaciones_basicas(3,2)
```

```python
suma, resta, producto, division = operaciones_basicas(3,2)
print("Suma: ", suma,"Resta: ", resta,"Producto: ", producto, "Division: ", division)
```

```python
def operaciones_basicas(a,b):
    suma = a+b
    resta = a-b
    producto = a*b
    division = a / b
    return {
        'suma':suma, 
        'resta':resta, 
        'producto':producto,
        'division':division
    }
    
```
