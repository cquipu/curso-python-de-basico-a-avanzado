```python
# lambda arguments: expression
suma = lambda a, b: a + b
```

```python
print(suma(5,6))
```

```python
## map, filter, sorted

numeros = [1,2,3,4,5]

cuadrados = list(map(lambda x: x**3, numeros))
print(cuadrados)
```

```python
numeros = [1,2,3,4,5,6]

pares = list(filter(lambda x: x%2 == 0, numeros))
print(pares)
```

```python
frutas = ['fresa','manzana','banana','cereza']

frutas_ordenadas = sorted(frutas, key = lambda fruta: len(fruta))
print(frutas_ordenadas)
```
