## Condicionales if, else, elif

```python
# Determinar si un numero es par

numero = 18

if numero % 2 == 0:
    print(f"{numero} es par")
else:
    print(f"{numero} no es par")
    
```

```python
# Determiniar si un numero es negativo, cero o positivo
numero = 0

if numero < 0:
    print(f"{numero} es negativo")
elif numero == 0:
    print(f"{numero} es cero")
else:
    print(f"{numero} es positivo")
```

## Match, case

```python
numero = 1
match numero:
    case 1:
        print("Opcion vegetariana")
    case 2:
        print("Opcion normal")
    case _:
        print("Ninguna opcion")

"""
match-case fue introducida en la version 3.10 de Python, en versiones anteriores no existe
"""
```

```python
numero = 6
match numero:
    case 0:
        print("Cero")
    case n if n > 0:
        print("Positivo")
    case n if n < 0:
        print("Negativo")
    case _:
        print("No se puede clasificar")

"""
la condicion:  n if n > 0 es una funcion anonima que recibe el parametro n 
"""
```

```python
# Determinar el Indice de masa corporal
imc = 30

match imc:
    case valor if 0 < valor <= 18.5:
        print("Bajo peso")
    case valor if 18.5 < valor <= 24.9:
        print("Peso normal (saludable)")
    case valor if 24.9 < valor <= 29.9:
        print("Sobrepeso")
    case valor if 29.9 < valor <= 34.9:
        print("Obesidad clase 1 (Moderada)")
    case valor if 34.9 < valor <= 39.9:
        print("Obesidad clase 2 (Severa)")
    case valor if 39.9 < valor:
        print("Obesidad clase 3 (Muy Severa)")
    case _:
        print("Valores del IMC no validos")

        
```

## for, break, continue

```python
secuencia = [1,4,5,2,3]
for elemento in secuencia:
    print(elemento)
```

```python
for letra in "Python":
    print (letra)
```

```python
range(4) #iterador
```

```python
for x in range(4):
    print(x)
```

```python
lista = [1,4,5,2,3]
for i in range(len(lista)):
    print(i)
```

```python
for indice, valor in enumerate(lista):
    print(f"Elemento {indice}: {valor}")
```

```python
secuencia = [1,4,5,2,3]
for elemento in secuencia:
    print(elemento)
    if elemento == 5:
        break
```

```python
secuencia = [1,4,5,2,3]
for elemento in secuencia:
    if elemento == 5:
        continue
    print(elemento)
```

# While

```python
contador = 0 
while contador < 5:
    print(f"Contador: {contador}")
    contador += 1

```

```python
contador = 0 
while True:
    print(f"Contador: {contador}")
    contador += 1
    if contador == 5:
        break
```

```python
contador = 0 
while contador < 5:    
    contador += 1
    
    if contador == 3:
        continue

    print(f"Contador: {contador}")
```

```python
def mi_funcion(par1, par2):
    res = par1 + par2
    return res
```

```python
def saludar(nombre, saludo):
    print(f"{saludo}, {nombre}!")

saludar("Juan","Hola")
```

```python
def saludar(nombre, saludo = "Hola"):
    print(f"{saludo}, {nombre}!")

saludar("Ana")
```

```python
def cuadrado(x):
    return x**2

resul = cuadrado(5)
resul
```

```python
x = 10 # variable global

def mi_funcion():
    y = 5 # variable local
    print(x + y)

mi_funcion()
```
