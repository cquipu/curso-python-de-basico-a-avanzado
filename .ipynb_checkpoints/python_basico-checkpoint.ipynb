## Condicionales if, else, elif

```python
# Determinar si un numero es par

numero = 18

if numero % 2 == 0:
    print(f"{numero} es par")
else:
    print(f"{numero} no es par")
    
```

```python
# Determiniar si un numero es negativo, cero o positivo
numero = 0

if numero < 0:
    print(f"{numero} es negativo")
elif numero == 0:
    print(f"{numero} es cero")
else:
    print(f"{numero} es positivo")
```

## Match, case

```python
numero = 1
match numero:
    case 1:
        print("Opcion vegetariana")
    case 2:
        print("Opcion normal")
    case _:
        print("Ninguna opcion")

"""
match-case fue introducida en la version 3.10 de Python, en versiones anteriores no existe
"""
```

```python
numero = 6
match numero:
    case 0:
        print("Cero")
    case n if n > 0:
        print("Positivo")
    case n if n < 0:
        print("Negativo")
    case _:
        print("No se puede clasificar")

"""
la condicion:  n if n > 0 es una funcion anonima que recibe el parametro n 
"""
```

```python
# Determinar el Indice de masa corporal
imc = 30

match imc:
    case valor if 0 < valor <= 18.5:
        print("Bajo peso")
    case valor if 18.5 < valor <= 24.9:
        print("Peso normal (saludable)")
    case valor if 24.9 < valor <= 29.9:
        print("Sobrepeso")
    case valor if 29.9 < valor <= 34.9:
        print("Obesidad clase 1 (Moderada)")
    case valor if 34.9 < valor <= 39.9:
        print("Obesidad clase 2 (Severa)")
    case valor if 39.9 < valor:
        print("Obesidad clase 3 (Muy Severa)")
    case _:
        print("Valores del IMC no validos")

        
```
